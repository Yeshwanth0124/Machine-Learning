```python
import os
import cv2
import numpy as np
import zipfile

with zipfile.ZipFile('train.zip','r') as zipfile:
  zipfile.extractall('train')
```

```python
root_folder = '/content/train/train'
```

```python
class_images = []
class_labels = []

data_labels = ['cats','dogs']
for labels in data_labels:
  folder_path = os.path.join(root_folder,labels)
  for filename in os.listdir(folder_path):
    if filename.endswith('.jpg'):
      image_path = os.path.join(folder_path,filename)
      image = cv2.imread(image_path)

      image = cv2.resize(image,(299,299))
      image = image/225.0
      class_images.append(image)
      class_labels.append(data_labels.index(labels))
images = np.array(class_images)
labels = np.array(class_labels)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(images,labels,test_size=0.2,random_state=42)
```

```python
X_train.shape ,  y_train.shape
```

```python
import tensorflow
from tensorflow import keras
```

```python
from keras.layers import Conv2D,MaxPooling2D,Flatten
```

```python
model = keras.Sequential([
    keras.layers.Dense(445,activation='relu',input_shape=(299,299,3)),
    keras.layers.Conv2D(32,(3,3),activation='relu'),
    keras.layers.MaxPooling2D((2,2)),
    keras.layers.Flatten(),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=10,batch_size=32,validation_data=(X_train,y_train))
```
